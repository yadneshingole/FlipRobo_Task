```python
Person_list=[]
Rank_list=[]
Net_worth_list=[]
Country_list=[]
Source_list=[]
Industry_list=[]

for i in urls:
    i.click()
       
    #Person name    
    name=driver.find_elements_by_xpath("//div[@class='personName']")
    for i in name:
        Person_list.append(i.text)
    #Rank
    rank_tag=driver.find_elements_by_xpath("//div[@class='rank']")
    for i in rank_tag:
        Rank_list.append(i.text)
    
    #Net worth
    net_worth=driver.find_elements_by_xpath("//div[@class='netWorth']")
    for i in net_worth:
        Net_worth_list.append(i.text)
    
    #Country
    country=driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
    for i in country:
        Country_list.append(i.text)
    
    #Source
    source=driver.find_elements_by_xpath("//div[@class='source-column']")
    for i in source:
        Source_list.append(i.text)
    
    #Industry
    industry=driver.find_elements_by_xpath("//div[@class='category']")
    for i in industry:
        Industry_list.append(i.text)
```

```python
#Checking length of list
print(len(Person_list),
len(Rank_list),
len(Net_worth_list),
len(Country_list),
len(Source_list),
len(Industry_list))
```

```python
# since the lenght of each lists are equal we will store in dataframe
import numpy as np
df=pd.DataFrame({})
df['Rank']=Rank_list
df['Name']=Person_list
df['Net Worth']=Net_worth_list
df['Country/Territory']=Country_list
df['Source']=Source_list
df['Industry']=Industry_list

df.index=np.arange(1,len(Rank_list)+1)
df
```
